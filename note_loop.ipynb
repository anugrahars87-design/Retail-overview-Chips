```python
#if msg send no need to send 3 times
succesfull=True
for num in range(3):
    print('good morning')
    if succesfull:
        print('succesfull')
        break #break the loop otherwise it print the same 2 msg 3 times
```

```python
#msg not sende in first
succesfull=False
for i in range(3):
    print('good morning')
    if succesfull:
        print("succesfull")
        break
```

```python
#msg not sende after 3 attemt so another msg to say that
succesfull=False
for num in range(3):
    print('good morning')
    if succesfull:
        print("succesfull")
        break
else:# doesnt need indentation if it came after the indentation, 4th msg will excute 3 times with the loop
    print("msg is not sended 3 times")
```

loop inside another loop

```python
for x in range(5):     # Outer loop: x goes from 0,1,2,3,4
    for y in range(3):   # inner loop: y goes from 0,1,2
        print(f"{x},{y}")
```
