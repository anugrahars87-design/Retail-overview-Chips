```python
#how to create our own custom function 
def greet():
    print("hai")
    print("anugraha")
greet()
```

```python
#aruments
def greet(first_name,last_name): #parameter : input that define for function
    print(f"{first_name} {last_name}")
    print("welcome")  
greet( 'anu','RS')    #argument :actual value for the given parameter
#greet( 'rihana' , 'pall')
```

```python
def greet(name):
    return(f" hi {name}")
print(greet('rosh'))
```

```python
def increment(num,by):
    return(num+by)
print(increment(2,10))
```

```python
def increment(num,by=2): #optional parameter comes last of requred parameter
    return(num+by)
print(increment(2))
```

```python
#we want products of given number
def multiply(x,y):
    return x*y
print(multiply(5,5))
```

more than 2 arguments


```python
  #  * for multiple arguments
def multiply(*numbers):  
    print(numbers)
multiply(5, 5, 6, 1, 2, 3)
```

```python
#multiply more than two arguments
def multiply(*numbers): 
 for num in numbers:
    print(num) #printing one at a time
multiply(5, 5, 6, 1, 2, 3)
```

```python
def multiply(*numbers):
 total=1            #create a variable initialize it 1 for multipliation,for adding it maybe 0 
 for num in numbers:
  total*=num   #total=total+num...for each iteration
 return total #after itertion return total   
print(multiply(5, 5, 6, 1, 2, 3))
```

```python
def divisible(inp):
    if inp %3==0:
        return 'fizz'
    if inp %5 ==0:
        return 'buzz'
    if (inp %3==0) and (inp %5 ==0):
        return 'fizzbuzz'
    return inp
print(divisible(15))
```
