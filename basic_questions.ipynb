 1)Reverse a string without using built-in functions

```python
def reversed_cha(s):    # Define a function, to reverse a string
  reversed_cha=""       #Start with an empty string    
  for i in s:           #Loop through each character in the input string
    reversed_cha= i + reversed_cha  #a,n+a,u+na,... this  formula is on revesed_char so when we put the text in it, it revese
  return reversed_cha               #    # Return the fully reversed string
text="anugraha"
result=reversed_cha(text)
print(result)
```

3)Find the factorial of a number (using loop & recursion)

```python
#using loop
def fact(n):
 result=1      #  Initialize result as 1 ,and it serves as the starting point for multiplication)
 for i in  range(1,n+1): # if n = 5,then (1,6)  then i = 1, 2, 3, 4, 5
    result*=i    # Multiply the current result by i, This accumulates the factorial step by step(result=result*1) i = 1 → result = 1 × 1 = 1,, i = 2 → result = 1 × 2 = 2,  2 × 3 = 6
 return result  #result holds the factorial of n

num=5
print(f"{fact(num)}")

    
```

```python
#using recurssion            Recursion is a programming technique where a function calls itself to solve
def fact_rec(n):
  if n==0 or n==1:           # if n is 0 or 1, return 1(because 0! = 1 and 1! = 1 by definition)
   return 1
  else:
   return n * fact_rec(n-1)  #- n * (n - 1) only multiplies two numbers once give 5*4 only but using fact_rec funtion calling -  rest of the multiplication (3 * 2 * 1)
num=5
print(f"{fact_rec(num)}")
```

4)Count vowels in a string

```python
def vow(s):
    vowels="aeiouAEIOU"  #vowels (both lowercase and uppercase)
    result=0             #Initialize result as 0  ,This will keep track of how many vowels we find
    for char in s:       ## Loop through each character in the input string
     if char in vowels:   # Check if the character is a vowel
      result+=1            # If yes, increment the counter by 1
    return result           # Return the total count of vowels found
text="hello world"
print(f"{vow(text)}")
```

5)Find the largest number in a list

```python
#simple
number=[1,3,5,6,4]
largest=max(number)
print(f"the largest num is {largest}")
```

```python
#by Loop
number=[1,2,4,6,7]
largest=number[0]    # Assume the first element is the largest initially
for num in number:   # Loop through each element in the list
   if num>largest:    # Compare current element with the current largest
     largest=num       # Update 'largest' if a bigger number is found
print(f"the largest num is {largest}")
      
```

6)Remove duplicates from a list

```python
numbers=[7,5,5,54,44,44,6,7,7]
unique = list(set(numbers)) #to remove duplicate
print(unique)
```

```python
number=[4, 4, 2, 2, 4, 67, 8]
unique=[]                   # Create an empty list to store unique values
for num in number:          # Loop through each element in the list 'number'
    if num not in unique:   # Check if the current element is already in 'unique'
        unique.append(num)   # If not, add it to 'unique
print(unique)
```

7)Swap two numbers without using a third variable

```python
#wrong way
a=5
b=10
print("a=",b,"b=",a) #will only swap the values at the moment of printing. It doesn’t actually change the variables themselve
```

```python
a=5
b=10
a=a+b   # a = 5 + 10 = 15 ,“hide” both numbers inside a
b=a-b   # b = 15 - 10 = 5 ,“extract” the old value of a into b.
a=a-b   #a = 15 - 5 = 10  ,“extract” the old value of b into a.
print("a=",a,"b=",b)  #a now holds the original value of b (10).b now holds the original value of a (5).
```
